## Probability

The study of statistics means we look at how probable of multiple events and explain them. For example, to see how happy is the person, we count how often the user include positive words in their scripts. This is interpreted as the probability of the user expresses positive words. 

In interpreting probability, we can map the values into language. The table below interprets what different probability values mean. 

| __Event__ | Impossible | Unlikely  | 50-50  |        Likely            | Certain |
|-----------|------------|-----------|--------|--------------------------|---------|
|    $p$    |  0         | < 50%     | 50%    |  > 50% and < 100%        | 100%    |

Probability can range between $0\%$ to $100\%$. While the earlier means it will never happen, the latter means it will certainly happen. 

In programming we convert probability values into decimals. For example, while $0\%$ is $0$, $100\%$ becomes $1$. You can think about if you have $100$ eggs and it is certain that any eggs you touched will break. You will can multiply $100$ with $1$ to explain how many eggs you expect to break. In the context of these exercises, we will use these decimals to represent probabilities.

## pdf/ cdf

How do we put the probabilities of events together? We order the probabilities with their random variables. This becomes a probability function of random variables, or simply __probability density function__ (pdf). It is called probability mass (pmf) function if the random variables are discrete numbers (integers). 

Sometimes we wanted to know the probability from a range of random variables, then we need from __cumulative distribution function__ (cdf). This is simply summing or integration of the range of random variables of interest. Another reason why cdf is important is where it only goes up, then the $y$-value of cdf is ordered in the sense each probabilistic value occured once. This gives way to another statistics called ordered statistics, which is out of scope of this course. Remember, random variables in this case are the $x$-axis. 

### Common pdf/ pmf

The common pdfs or pfms are: 
* Uniform distribution
* Binomial distribution
* Poisson distribution
* Chi-squared distribution
* Normal distribution

Uniform distribution means all events occured with the same probability. The pdf is 
$$f(x)\;=\;\frac{1}{b\;-\;a}\mathbb{I}_{a\;\leq\;x\;\leq\;b}$$

The function itself does not depend on any input random variables $x$. So this means the function looks like a horizontal line between $a$ and $b$. For example, if the probability of making eggs or making cereal as breakfast are 50-50, their probability are the same. It's pdf looks like 
$$f(x)\;=\;\frac{1}{2}\mathbb{I}_{0\;\leq\;x\;\leq\;2}$$

````python
# Try me
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

x = np.linspace(0,2,100)
rv01 = stats.uniform.pdf(x, 0, 2)
plt.plot(x, rv01)
plt.xlim(-0.5, 2.5)
plt.show()
````

Binomial distribution $\operatorname{Bin}(n,\:p)$ models numbers of response of "yes" or "no" and how probable they appear each time. The parameters are: 
* $n$ is the total number of responses. 
* $p$ is the probability of event happening (for example "yes"). 
The mean in binomial distribution is $np$. It's pdf is 

$$f(x)\;=\;{n\choose k}p^k q^{n\;-\;k}$$

In this example we look at when $\operatorname{Bin}(10,\:0.5)$. 

````python
# Try me
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

x = np.arange(0,11)
rv02 = [stats.binom.pmf(i, 10, 0.5) for i in x]  # pmf is for discrete random variables 
plt.plot(x, rv02)
plt.xlim(-0.05, 10.05)
plt.show()
````

We can see from here, the peak of probability touches $0.25$ if we are looking for $5$ events occured. The reason why this doesn't touches the parameter $p$ is the probability parameter specifies for __each__ time, how often will it appear. If they come in series, then the probability of having them appearing is less than the individual probability.

Linear transformation means if we add a constant, or scale them by a constant. By adding different columns of data also part of linear transformation (unless if it not row by row). 

You may find this helpful next week (or in week 10). The mean and standard deviation of linearly transformed data is 
$\operatorname{E}(aX\;+\;b)\;=\;a\operatorname{E}(X)\;+\;b$
The mean of the transformed data is simply done by transforming the mean value as the same. For example, degrees of Celsius to Fahrenheit is defined by $F\;=\;\frac{9}{5}C\;+\;32$. Let's say we have the measured temperature of several chemical samples and trying to find its mean temperature. If the mean temperature in degrees of Celsius is $11^\circ\mathrm{C}$, then the mean temperature in Fahrenheit is $\frac{9}{5}\;\times\;11\;+\;32\;=\;51.8^\circ\mathrm{F}$. 

The variance of the linearly transformed data is 
$\operatorname{Var}(aX\;+\;b)\;=\;a^2\operatorname{Var}(X)$
Where the additional constant has no use in the transformation. 

## Experimental Design (Optional)

If you are from biological studies, humanities or related, experimental design is an important. This involves planning on studying the research question by __testing if the outcome/ results are expected under a range of factors__. The most basic principle of experimental design is we tested several independent variables and measure the dependent variables (outcome), while controlling other variables. 

Let us use an example, suppose we are looking how long can an object fall at different heights. So we are using height as independent variables, time to fall for dependent variable. At the same time we would want to control the room temperature, pressure. These are to prevent any external impact on our experiments, since often we proceed each experiment sets in series. More prominently, if we control the object to be experimented, which maintains the same among all sets, then it would be a better control factor. 

| Independent variable(s) | Dependent variable(s) | Control variables            |
|-------------------------|-----------------------|------------------------------|
| Height                  | Time to fall          | Object <br> Room temperature |

In this section, we use the term __factor__ to represent each of these variables. For example, height is a factor, so as time to fall. Each factor allows a range of values and they are called __level__. 

Experimental design has the three main principle: __Control__, __randomisation__ and __replication__. Let us see what do they mean in below: 
* __Control groups are important in eliminating systematic errors and external effects from the experiment__. This means as an experimenter, we need to be clear of what are we measuring, and other possible factors are potential control factors. The most easy way is to ensure the control factor has one value among all experiment sets. __Blocking sets__ are designed for this when other external factor may apply. The experimenter will select one blocking factor at a time in the experiment, and measure the experiment as usual. The experiment sets that uses the same blocking factor, as a block. Experimenters may compare the effect between blocks and see how the external factor affects the results. 
* __Randomisation is another way to eliminate bias__. In simple experiments like finding the time to fall on ground, there are not much factors to play around. Often in clinical trials, there are external factors that cannot be controlled or observed. For example, demographics. They may impact to the results, and randomisation means either we assign them into groups and the experiment factors are randomly assigned (i.e. providing real medication or placebo, or apply treatment in a radomly permutated order). By such, we eliminated the effect from demographics (i.e. by assigning different demographics to same group we can see how the general effect to the medication) or subjective judgement (i.e. experimenters may have done differently which reduce effectiveness of trial). Perfect randomisation cannot be achieved but we rely on randomised design to improve this. This is discussed in next (sub)section. 
* __Replicating the experimental procedure produces results in a condient manner__. For example, if we are trying to see how many red or black balls in a bag, the best way is to take out the balls one-by-one and we conclude by finding how often we take a red or black ball on average. Experiments find results from unknown, but we can still see the expected value by replicating the sets again and again. 

We may combine and create the structure of the factors in experimenting them. 

* __$2^k$ Factorial__ This means we have a set of factors that accepts only binary levels. The most easy examples are "yes" or "no", "male" or "female". This is very common in life. 
* __Fractional Factorial__ It means we screen for which factor are important and we will experiment a subset of factors. 
* __$3^k$ Factorial__ An extension to the $2^k$ factorial, but this time each factors has $3$ responses. 
* __Nested__ The factors formed a hierarchical structure. The levels of the same factor does not always combine with all the level in the next factor, they might select a subset of them. 
* __Crossed__ The factors fully combined with each other. 
* __Unbalanced Designs__ This simply means some of the combination of levels has more than one observed results. Usually it's because of practical reasons that make unbalanced designs possible (e.g. balanced design produce unbalanced data size when things go wrong, too expensive, some of the level combinations are more of the interest).

### Randomised Block Design (Optional)

When one considers experimenting with multiple factors, they should ensure any systemic bias are eliminated. Usual procedure is to compare the factors between different block designs. Depends on the structure of factors, there are many ways to create block designs. Here are an exhausted list of them: 

* __Randomised Complete Block Design__ This simply means all levels crossed over, and each cross over (namely the experiment set) is generated in a random order. The blocks are introduced as the homogeneous iteration of this randomised experiment design. This design generates very small experimental error, if not the least error among others. 
* __Latin Squares__ If there are two blocking factors, then we may align them as a latin square and arrange the blocks (and experimented level combination) within the cells. The examples are listed below the dot points. 
* __Incomplete Block__ Compared to complete block, not all factor levels are tested. The experimenter is interested between the difference of some of the factors, and the different outcomes from these. 
* __Split-plot__ Arrange different experiment sets into separate experiment plots, or equipments or blocking factors. This means each of these tests several experiment sets but they don't span the whole complete block. The nested and crossed has the form this above.

### Bootstrap

From statistics we know that the more data we have, the more likely they show an actual answer from their means. When we don't have many data then we can "make some". Bootstrapping is to produce data from iterating the random sampling from the dataset again and again. So some samples overlap each other, but the combination in the samples will increase the confidence of the statistic. 

The below code shows a demo of bootstrapping. 

````python
# Run me
import numpy as np

bootstrap_means = []
runs = 100

for _ in range(runs): 
    bootstrap_sample = df.sample(df.shape[0]//runs)  # Random sample of each iteration
    bootstrap_mean = bootstrap_sample['Age'].mean()
    bootstrap_means.append(bootstrap_mean)
    
bootstrap_means = np.array(bootstrap_means)
bootstrap_means
````

The mean from bootstrapping is 

````python
bootstrap_means.mean()
````

````python
df['Age'].mean()
````

Compared to the actual mean, which is not too much far from it. 

We can also look at the confidence interval of the mean age from boostrapping. 

````python
np.percentile(bootstrap_means, [2.5, 97.5])
````

It means that $95\%$ of the time that the sampled means fall between 24.5475 and 32.305. This importance becomes apparent because datasets measures a subset of the whole population. When there is more data, we can see where the new mean value may fall, likely within the confidence interval and that is how we estimate the mean age of respondants. 

## Hypothesis Testing

When we do conduct an experiment, or aggregating survey results. We always want to infer the conclusion from the large amount of data. Often when we obtain the results, we would want to make sure this is not due to systematic measurement errors or statistical fluctuations. __Hypothesis testing__ is aimed for testing the hypothesis and whether they are statistically significant to claim the conclusion. 

There are few ingredients to conduct a hypothesis testing, this depends on what are you trying to test. 
* A __null hypothesis__ that defines what we aimed to test of. For example, if you wanted to test if the sample mean if equal to $3$. Note that a null hypothesis has to be quantitative, or at least it can be related to a quantitative statement. The complement of a null hypothesis is the __alternative hypothesis__. If we reject the null hypothesis, that means the alternative hypothesis has some statistical significance in it. 
* __test statistic__ is the quantitative object that we used in the hypothesis testing. The hypothesis testing method uses a particular pdf and this is their variable. 
* We then set a __significance level__ to see how how much significance we need. The consensus is the signficance level above 95% ($\alpha\;=\;0.05$) and we accept the null hypothesis. 

In hypothesis testing, one way is to see if the measured $p$-value is less than the complement of the significance level. For example, if the significance level falls on 95%, then if the $p$-value is less than $1\;-\;0.95\;=\;0.05$ then we reject the null hypothesis. The more advanced way is to look if the test statistic falls inside the rejection region. 

| Truth/ action       | $H_0$ is true | $H_0$ is false               |
|---------------------|---------------|------------------------------|
| Reject $H_0$        | __Type I__    | Correct Power: $1\;-\;\beta$ |
| Not rejecting $H_0$ | Correct       | __Type II__                  |

### Proportion Testing

We would wish to see if there is a proportion of samples have the same property. For example, if we wish to see if more than half are male. Then we use proportion testing. In here we want to test, 

$H_0:\:p\;=\;0.5 $ (null hypothesis)

$H_A:\:p\;\neq\;0.5$ (alternate hypothesis)

This type of hypothesis testing uses z-score as test statistic, which is defined as 

$$z\;=\;\frac{\hat{p}\;-\;p\text{*}}{\sqrt{\frac{p\text{*}\left(1\;-\;p\text{*}\right)}{n}}}$$
Where $p\text{*}$ is $0.5$ in this case. Let us say there are $25$ survey samples where $13$ of them are responded by a male. So we can get the z-score equals to $z\;=\;0.748$. Depends of whether you use $p$-value or critical (rejection) region. Here are the next steps: 
* Use a conversion table to see where the z-score converts to the corresponding $p$-value. Compare if it is less than the critical value (commonly $<\;0.05$ for 95% significance). 
* Find the critical region (i.e. cutoff of z-score) from the conversion table (to $p$-value) and check if the z-score is within the critical region. 

Hence we have 
* The corresponding $p$-value is $p\;=\;0.454$ (Note this $p$ is not the $p$ in the null hypothesis). Which is larger than the critical value. 
* The critical region lies between $z\;=\;-1.96$ and $z\;=\;1.96$. The measured z-score is within the critical region. 

The two methods shows that we should accept the null hypothesis. 

__Hypothesis for Two-tailed test__

$H_0:\:p\;=\;p\text{*} $

$H_A:\:p\;\neq\;p\text{*}$

The critical region is within the critical value $\pm z\text{*}$. 

__Hypothesis for Upper-tailed test__

$H_0:\:p\;=\;p\text{*} $

$H_A:\:p\;>\;p\text{*}$

The critical region is less than the critical value $z\text{*}$. 

__Hypothesis for Lower-tailed test__

$H_0:\:p\;=\;p\text{*} $

$H_A:\:p\;<\;p\text{*}$

The critical region is more than the critical value $-z\text{*}$. 

In Python we have `proportions_ztest()` from `statsmodels` to conduct a proportion test. The following code will demonstrate from the example above. 

````python
# Run this code 
from statsmodels.stats.proportion import proportions_ztest

'''
Usage: 

statsmodels.stats.proportion.proportions_ztest(count, nobs, value=None, alternative='two-sided', prop_var=False)
'''

z, p = proportions_ztest(13, 25, value=0.5, alternative='two-sided')
````

The function outputs two objects: The z-score and the p-value. 

````python
# Run this code 
print(f'z = {z}')  # s-score
print(f'p = {p}')  # p-value 
````

### t-test

t-test tests the sample mean, and if they are big enough, small enough or equal to a reference value. It uses a test statistic called t-score. It is defined as 
$$t\;=\;\frac{\bar{x}\;-\;\mu_0}{\frac{s}{\sqrt{n}}}$$

First, you need to know which pair of hypothesis is needed, the following is as a guide. 

__Hypothesis ($t$-test)__

$H_0:\:\mu\;=\;m$

$H_A:\:\mu\;\neq\;\mu\text{*}$ (Two tail)

$H_A:\:\mu\;<\;\mu\text{*}$ (Lower tail)

$H_A:\:\mu\;>\;\mu\text{*}$ (Upper tail)

For example, if you wish to test whether the sample mean is less than $99$. The hypothesis reads 

$H_0:\:\mu\;=\;99$

$H_A:\:\mu\;<\;99$ 

The null hypothesis is always in this form (i.e. with equal), and our goal is to see whether we reject the null hypothesis. 

Depends of whether you use $p$-value or critical (rejection) region. Here are the next steps: 
* Use a conversion table to see where the t-score converts to the corresponding $p$-value. Compare if it is less than the critical value (commonly $<\;0.05$ for 95% significance). 
* Find the critical region (i.e. cutoff of z-score) from the conversion table (to $p$-value) and check if the t-score is within the critical region. 

In Python we have `ttest_ind()` from `statsmodels.stats.weightstats` to calculate the t-test. The following code will demonstrate this. For example, if we want to see if a sample data has mean less than 99, then 

````python
# Run this 
'''
scipy.stats.ttest_1samp(a, popmean, axis=0, nan_policy='propagate')
'''
import numpy as np
from scipy import stats

np.random.seed(1000)  # fix seed to get the same result
data_ttest01 = stats.norm.rvs(loc=105, scale=10, size=100)
t, p = stats.ttest_1samp(data_ttest01, 99)
print(f't = {t}')  # t-score
print(f'p = {p}')  # p-value 
````

As you can see from the $p$-value the value is very small and we can conclude there is a __statistically significant__ evidence where the two distributions does not have the same mean. 

There are different types of t-tests, this depends on if you are comparing the mean values between samples to a value or between samples. They are 
* One sample t-test
* Independent two-sample t-test
* Paired sample t-test

The one above is the one sample t-test. 

If we have two datasets and which to compare their sample means. We can use two-sample t-tests. In Python we have 

* `ttest_ind()` from `statsmodels.stats.weightstats`, or 
* `ttest_ind()` from `scipy.stats`

to conduct an independent two-sample t-test. The following code will demonstrate from the example above using `statsmodels` library.

### Chi-square test

Chi-square test is used for testing if the categorical data are dependent (or not). Categorical data could be, for example, male and female. We test of the data collected from either categories are dependent to each other. 

The null hypothesis of chi-square test is 

__Hypothesis__

$H_0:\:$ Data are independent. 

$H_A:\:$ Data are not independent. 

As a reminder, if the $p$-value is lower than the critical value then we reject the null hypothesis. 

We can use `chi2_contingency()` from `scipy` to test whether each category differs. The method accepts one required argument, which is a two-dimensional table of observed frequencies (i.e. number of occurrences) in each category. (It could ne multi-dimensional but we often not use this) In the following example, we have $2$ random variables from each categories (hence and 2-by-2 table): 
```python
[[10,21],
 [28,5]]
```
Which looks like this if we handraw them: 

|.      | $X_1$ | $X_2$ |
|-------|-------|-------|
| $Y_1$ |   10  |   21  |
| $Y_2$ |   28  |   5   |

You can run the following code to see what is happening 

````python
# Run this 
'''
scipy.stats.chi2_contingency(observed, correction=True, lambda_=None)
'''
from scipy.stats import chi2_contingency

chi1, p, df, expected = chi2_contingency([[10,21],[28,5]])
print(chi1, p, df, expected)
````

From the results, we can clearly see how small of the $p$-value. It shows that the counts from each category are independent to each other. The method outputs other objects: 
* `chi2` is the test statistic from the test, we use this if we follow the rejection region method. 
* `df` is the degrees of freedom, in this case it is number of categories minus one. 
* `expected` is the (contingency) table if the data are independent to each other. 

## Statistical Power Testing (Optional)

Right now we have looked at the different types of hypothesis testing. These hypothesis testing involves our data forms as a distribution and we compare with another distribution tha the test offers. Specifically this means we compare the, for example, the mean or other parameters, so that we can decide if the distributions are different. 

Because we are comparing with a theoretical distribution with our data. If the number of data is very small, often the estimate the mean (for example) has not converged to its true value yet. In this case, we may end up __not rejecting__ the null hypothesis incorrectly. So we need statistical power tests to determine how many data points is enough for us to determine if we reject the null hypothesis. 

* There is a term called __sample size__. This happens when our data set comes from multiple experiments and this term is the sample size of how many data points observed in each experiment. In case when the data comes from given datasets, then we do not know what the sample size, then we can use the whole dataset as the sample size. 
* In power analysis, we aggregate the mean and standard deviation to compare the distributions. This combination of mean and standard deviation (or other parameters of the distribution) would generate __effect size__. There is no one set of way to calculate this, it could be calculated by any statistical calculator (google them and you will find couple) or even education guess. 

The quantitative measure of not incorrectly rejecting the null hypothesis is the __power__. Like the confidence level $\alpha$, we have a common criteria of higher than 0.8 means we have sufficient power. The end goal of statistical power testing is to find out minimum data size based on the statistical power $1\;-\;\beta$, effect size and sample size. 

In `statsmodel` there are couple of methods that possess statistical power testing. You may look at [here](https://www.statsmodels.org/stable/stats.html#power-and-sample-size-calculations) for the list of methods offered by `statsmodel`. 

In this week we will use from the one-sample t-test from this section. The following code involves preparaing the dataset (or simply run the original one above and call it in the cell below), effect size using Cohen's $d$ and run the test. 

````python
# Try me
import statsmodels as sm

data_01 = stats.norm.rvs(loc=5,scale=10,size=500)
d = (data_01.mean() - 99)/data_01.std()
print(d)

obj = sm.stats.power.TTestIndPower()
obj.power(effect_size=d, nobs1=500, alpha=0.05, alternative='smaller')
````

Since the power is greater than the criteria 0.8, we could say the the probability that we reject the null hypothesis falsely is minimal. This means with 500 data points, we have found statistically significant evidence that we have rejected the null hypothesis correctly. 

`statsmodel` also provides a way to visualise and decide how many data points will lead to satisfied power using `.plot_power()`. If we set the `dep_var` as `'nobs'`, it means we use the number of data as $x$ axis. The method accepts ranges of sample and effect sizes and visualise them with the power computed. 

````python
# Try me
effect_sizes = np.array([1, 10])  # Testing different effect sizes
sample_sizes = np.array(range(5, 100)) # Testing different sample sizes

obj.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes)
````

## Categorical Regression (Optional)

Catagorical variables unlike random variables, they do not necessarily follow an order (so it is meaningless to draw a regression line under male and female values). So special care has to be applied when we uses categorical variables in our analysis. Often this is done by coding. The common ones are __simple coding__ or __dummy coding__. 

Simple coding means we assign integer values to each catagories. For example, we may assign  

| Value  | $C_1$ | 
|--------|-------| 
| Female | 0     | 
| Male   | 1     | 

This means the regression slope shows how many quanities (e.g. salary earned) that male earned more/ less than female. We can extend this to more levels such as 

| Value  | $C_1$ | 
|--------|-------| 
| Female | 0     | 
| Male   | 1     | 
| Other  | 2     | 

Hence the same logic applies. For different gender categories, there is a uniform difference of quanities (e.g. salary earned). This means if $C_1 = 1.2$, that means men earn 1.2 more than female and other gender(s) earn 1.2 more than male. Gender is a nominal feature and often we do not assume the variables themselves mapped to ordered dependent varaibles (i.e. outcomes). So the next type of coding is what we can use when there are many levels. 

In Python we can encode the string categorical values using simple coding. Sometimes the dataset has already encoded them for you. 

Dummy coding is where we assign a flag to each catagory. For example, 

| Value  | $C_1$ | $C_2$ | $C_3$ |
|--------|-------|-------|-------|
| Male   | 0     | 1     | 0     |
| Female | 1     | 0     | 0     |
| Other  | 0     | 0     | 1     |

In this example, we have male/ female or other gender as categories, and they assign to each dummy variables. The linear regression equation looks like: 

$$\hat y\;=\;\hat\beta_1\:C_1\;+\;\hat\beta_2\:C_2\;+\;\hat\beta_3\:C_3$$
Remember the slope is something we ought to fit them next, so we should not worry about them. The slope means how each categorical variables effect the outcome, and its significance.

### Using Logistic Regression

Logistic regression maps all factors into probability in fact, not $0$s and $1$s. This is something you should be aware when using logistic regression. 

Logistic regression use probabilistic numbers as its dependent variable. By rearranging the regression equation, we have 

$$p = \frac{1}{1\,+\,\mathrm{e}^{-\left(\hat{\beta}_0\;+\;\hat{\beta}_1 x_1\;+\;\cdots\;+\;\hat{\beta}_n x_n \right)}}$$

By convention we say more than $0.5$ means we assign the label $1$ and vice versa. 

Since logistic regression is to predict whether something happens. Like betting we introduce the term odds which means how likely the event may happen. Which is defined by 

$$\text{odds} = \mathrm{e}^{\hat{\beta}_0\;+\;\hat{\beta}_1 x_1\;+\;\cdots\;+\;\hat{\beta}_n x_n}$$

Unlike probability this can be any number. The larger the odds, the likely it will happen. But how can we say with just one number? Of course we compare with the alternative event through its ratio. So the event you wish is on the numerator, and the others in the denominator, then when the odds ratio is less than one then the event is said to be unlikely to be happened. 

Odds ratio is defined as 

$$\frac{\text{odds}(\text{1})}{\text{odds}(\text{0})}$$

## ANOVA

ANOVA is used for understanding whether a factor, for example, male or female in gender, if they produce different outcomes. For example, gender to salary. We use a F-statistic test to see whether: 

$H_0$: The means from each factor are the same. 

$H_A$: There is one factor produces mean observation signficantly differ to other factors. 

This time we will use the survey dataset from multiple regression and apply ANOVA in below: 

````python
# Get linear regression model
import pandas as pd
import statsmodels.api as sm

data_01 = df = pd.read_csv("files/survey_data.csv")
m1 = sm.formula.ols(formula="Salary ~ C(Gender) + Age",data=df).fit()
````

Note: This time we use something called `patsy` formula syntax to construct the linear regression equation. If you not have done the categorical regression, the `C()` means it is a categorical parameter. 

So we process ANOVA with the `sm.stats.anova_lm()` function, with the fitted linear model as argument. 

````python
sm.stats.anova_lm(m1)
````

How can we capture this results? Apart from finding the mean salary from each gender. One way we can do is to visualise the different genders with boxplot and see where they stand.

From above we can see that the green line (box actually) represents where the box locates. They are located very similarly, except for men, few people earn more and the outlier explains why different gender may have earned different salary level. 

Of course, applying ANOVA in multiple factors is more useful: We can use this to explain if there is a particular factor differs or significant with others. This time we would add the different qualifications that the survey respondents hold. 

````python
m2 = sm.formula.ols(formula="Salary ~ C(Gender) + C(Qualifications) + Age",data=df).fit()
````

We then apply the boxplots to see which qualifications has a different effect than others. 

````python
data_01.boxplot(column="Salary", by="Qualifications")
````

But is there a way to check which factor differs with others without guessing from visualisation? We can use Tukey Honestly Significant Difference (HSD) test to check if the mean difference are significant. Reading Tukey test is the from reading a hypothesis test. It uses a statistic called 
$$\operatorname{HSD}\;=\;q_{k, df, \alpha}\sqrt{\frac{\operatorname{MSE}}{n}}$$
Where $q_{k, df, \alpha}$ is the critical value from studentised range distribution from number of levels $k$, degrees of freedon and the confidence level $\alpha$. Then multiply with the mean squred error and the number of data $n$. We use this statistic and compare with the mean distance between two factors. 

So we decide where 

$H_0$: The means from the two factors are the same. ($\vert \mu_i - \mu_j\vert\;<\;\operatorname{HSD}$)

$H_A$: There is one factor produces mean observation signficantly differ to other factor. ($\vert \mu_i - \mu_j\vert\;>\;\operatorname{HSD}$)

We can find out which factor(s) differ from others using the `sm.stats.multicomp.pairwise_tukeyhsd()` function. 

````python
tukey = sm.stats.multicomp.pairwise_tukeyhsd(endog=data_01['Salary'],
                          groups=df['Qualifications'],
                          alpha=0.05)
print(tukey)  # Jupyter does not have inline printing for this
````

As expected from the ANOVA results, no factors are significantly differ than others. So the Tukey test should tell the same: We see the last column says all pair of factors should accept the null hypothesis.

Akaike's Information Criterion (AIC) estimates the prediction error and used to describe the quality of the models: The lower AIC means less information losses through the model and hence we often choose the models with lower AIC. AIC is defined by

$$\operatorname{AIC}\;=\;2n\;-\;2\ln\left(\hat{\mathcal{L}}\right) $$

Where it is determined by the number of parameters $n$ and the log-likelihood $\ln\left(\hat{\mathcal{L}}\right)$ from the Estimate Parameters section above. 

We have already seen AIC before this in fact, this is shown in the linear regression summary statistics. This is in the right hand side of the table.

## Information Theory (Optional)

In this last section we will look at an application of probability, which is what does the events in the probability space themselves means as a whole. For example, if one event dominates or if other events might happen at a suprise? Information theory looks into that. 

Information comes from events and we capture them as probability. For example, in coin tosses we have half of the time which shows head and half of the time shows tail. So we have all probabilities from coin tossing as 
```python
prob = [0.5, 0.5]
```
Given one toss has two outcomes, so how much information does either outcome gives us? We use __information__ to answer that, which is 

\[I(X)\;=\;-\log_2 p\]

What this means is for the binary events, how much do we know? For example, we know that the probability of three heads is $\left(\frac{1}{2}\right)^3\;=\;\frac{1}{8}$. So $I(X=3)\;=\;3$ which means the information quantity has some relation to number of tosses. 

In Python we can calculate information with

### Entropy (Optional)

Entropy is the central measure of information theory. For example, we have an event that has several outcomes. Like coin tossing has probabilities of $0.5$ (head) or $0.5$ (tail). Entropy is the average of the information produced by each outcomes. 

For completeness, the equation for entropy is $H(X)\;=\;-\sum_X P(X)\log_2P(X)$, the code below computes this. 

````python
# Run me
import numpy as np
import math

def entropy(p_ls):
    return np.mean(np.array([-(x * math.log2(x)) for x in p_ls]))

entropy([0.1, 0.2])
````

Often people said entropy is a measure of suprises. This means higher entropy means there will be suprises from a rare outcome. High entropy also means there are variability from the measured outcomes, or accessible states. 

When we have different random variables, for example $X$ represents coin tosses and $Y$ represents outcomes from paper, scizor, rocks. Then we use joint entropy here. The equation is $H(X,Y)\;=\;-\sum_Y\sum_X P(X,Y)\log_2P(X,Y)$ and the code below computes joint entropy. 

````python
# Run me
import numpy as np
import math

def joint_entropy(p_ls):
    p_ls = np.array(p_ls)
    return np.sum(-p_ls * np.log2(p_ls))
#     return np.mean(np.array([-(p * math.log(p)) for x in p_ls]).flatten())

joint_entropy([[0.1, 0.2], [0.1, 0.5]])
````

The input parameters are what is called joint probability table. The first element is when $X\;=\;0$ and $Y\;=\;0$, and we change one random variable that represents each list elements. So the second element (first list) should be $X\;=\;1$ and $Y\;=\;0$. This is because we look for one different element horizontally. 

The other important measure comes if we assumed for some outcome from a random variable and the entropy from another one. In fact, we are conditioning one. This is the conditional entropy. 

The equation is $H(X\vert Y)\;=\;-\sum_Y\sum_X P(X,Y)\log_2\frac{P(X,Y)}{P(Y)}$

````python
# Run me
import numpy as np
import math

def cond_entropy(p_ls, rnd_var='Y'):
    if rnd_var == 'X': 
        p_x = np.sum(p_ls, axis=1)
    elif rnd_var == 'Y': 
        p_x = np.sum(p_ls, axis=0).reshape(2,1)
    return -np.sum(p_ls * np.log2(p_ls/p_x))

cond_entropy([[0.25, 0.25], [0.25, 0.25]])
````

We look at mutual information $I(X;Y)$. This measures how much does $X$ depends on $Y$. If they are inpdependent, then the mutual information is $0$. Higher the value, likely that when we know one outcome from $X$ or $Y$ happens from looking the probabilities from the other random variable. 

````python
# Run me
import numpy as np
import math

def mutual_info(p_ls): 
    p_x = np.repeat(np.sum(p_ls, axis=1).reshape(2,1),2,axis=1)
    p_y = np.tile(np.sum(p_ls, axis=0),(2,1))

    return np.sum(p_ls*np.log2(p_ls/(p_x*p_y)))

mutual_info([[0.25, 0.25], [0.25, 0.25]])
````

Finally, we look at a relatively important topic in machine learning. Which is called __KullbackLeibler divergence__ or relative entropy. It shows how much infomation gained if we used one distribution than the other. 

The following code shows probabilities from two random variables, with the top row as first distribution and so forth. 

````python
# Run me
import numpy as np
import math

def mutual_info(pq_ls): 
    for x in pq_ls: 
        if not (type(x) != list or type(x) != np.array): 
            raise TypeError("Input must be two dimensional array. ")
    if len(pq_ls) != 2: 
        raise ValueError(f"Input list must have row dimension of 2, current: {len(pq_ls)}")
    pq_ls = np.array(pq_ls)
    return np.sum(pq_ls[0]*np.log2(pq_ls[0]/pq_ls[1]))
    
mutual_info([[0.5, 0.3, 0.1, 0.1], [0.1, 0.25, 0.25, 0.4]])
````
